```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import savgol_filter

# seed
test_accuracy_seed0 = np.load('/home/s3963616/AML/results/nw5-nss1-nqs15-mbs1-vi500-net1000-nte40000-l0.001-il0.4-soFalse-domniglot-T1-is28-rFalse-dNone-s0/test-accuracy.npy')
test_accuracy_seed18 = np.load('/home/s3963616/AML/results/nw5-nss1-nqs15-mbs1-vi500-net1000-nte40000-l0.001-il0.4-soFalse-domniglot-T1-is28-rFalse-dcuda:0-s18/test-accuracy.npy')
test_accuracy_seed2017 = np.load('/home/s3963616/AML/results/nw5-nss1-nqs15-mbs1-vi500-net1000-nte40000-l0.001-il0.4-soFalse-domniglot-T1-is28-rFalse-dcuda:1-s2017/test-accuracy.npy')

test_accuracy_nw6_seed0 = np.load('/home/s3963616/AML/results/nw6-nss1-nqs15-mbs1-vi500-net1000-nte40000-l0.001-il0.4-soTrue-domniglot-T1-is28-rFalse-dcuda:0-s0/test-accuracy.npy')
test_accuracy_nw6_seed18 = np.load('/home/s3963616/AML/results/nw6-nss1-nqs15-mbs1-vi500-net1000-nte40000-l0.001-il0.4-soFalse-domniglot-T1-is28-rFalse-dcuda:1-s18/test-accuracy.npy')
test_accuracy_nw6_seed2017 = np.load('/home/s3963616/AML/results/nw6-nss1-nqs15-mbs1-vi500-net1000-nte40000-l0.001-il0.4-soFalse-domniglot-T1-is28-rFalse-dcuda:0-s2017/test-accuracy.npy')

# 
def smooth_data_savgol(data, window_length=51, polyorder=3):
    # 
    if window_length > len(data):
        window_length = len(data) // 2 * 2 + 1
    elif window_length % 2 == 0:
        window_length += 1
    return savgol_filter(data, window_length, polyorder)

# 
plt.figure(figsize=(12, 6))

# 
plt.plot(smooth_data_savgol(test_accuracy_seed0), label='5-ways Seed 0', alpha=0.8)
plt.plot(smooth_data_savgol(test_accuracy_seed18), label='5-ways Seed 18', alpha=0.8)
plt.plot(smooth_data_savgol(test_accuracy_seed2017), label='5-ways Seed 2017', alpha=0.8)

plt.plot(smooth_data_savgol(test_accuracy_nw6_seed0), label='6-ways Seed 0', alpha=0.8)
plt.plot(smooth_data_savgol(test_accuracy_nw6_seed18), label='6-ways Seed 18', alpha=0.8)
plt.plot(smooth_data_savgol(test_accuracy_nw6_seed2017), label='6-ways Seed 2017', alpha=0.8)

plt.title('Smoothed Test Accuracy Trend for 5-ways vs 6-ways Classification')
plt.xlabel('Interval Task Count')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
